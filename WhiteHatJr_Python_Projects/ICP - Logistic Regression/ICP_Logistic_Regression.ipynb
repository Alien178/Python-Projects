```python
import csv
import pandas as pd
import plotly.express as px

df = pd.read_csv("data.csv")
score = df["Score"].tolist()
accepted = df["Accepted"].tolist()

fig = px.scatter(x = score, y = accepted)
fig.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
import numpy as np

x = np.reshape(score, (len(score), 1))
y = np.reshape(accepted, (len(accepted), 1))

lr = LogisticRegression()
lr.fit(x, y)
plt.figure()
plt.scatter(x.ravel(), y, color = "black", zorder = 20)

def model(x):
  return 1 / (1 + np.exp(-x))

x_test = np.linspace(0, 100, 200)
chances = model(x_test * lr.coef_ + lr.intercept_).ravel()

plt.plot(x_test, chances, color = "red", linewidth = 3)

plt.axhline(y = 0, color = "black", linestyle = "-")
plt.axhline(y = 0.5, color = "green", linestyle = "--")
plt.axhline(y=1, color = "black", linestyle = "-")
plt.axvline(x = x_test[165], color = "blue", linestyle = "--")

plt.ylabel("Y")
plt.xlabel("X")

plt.xlim(75, 85)

plt.show()
```

```python
userscore = float(input("Enter your Marks: \n"))
chances = model(userscore * lr.coef_ + lr.intercept_)

if chances <= 0.01:
  print("You Failed")

elif chances >= 1:
  print("You Passed!!!!")

elif chances < 0.5:
  print("You might not pass")

else:
  print("You might pass")
```
